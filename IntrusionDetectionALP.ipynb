# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
for x in train_df.select_dtypes(include=['float64', 'int64']):
    train_df[x] = minmax.fit_transform(np.array(train_df[x]).reshape(-1,1))
```
